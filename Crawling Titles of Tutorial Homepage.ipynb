```python
import pandas as pd
import requests # for html 
from bs4 import BeautifulSoup
```

```python
def crawlTitle(get) :
    # get html
    hp = "https://entagma.com/courses/" 
    merge = hp + get + "/"
    res = requests.get(merge)

    #parse wanted part 
    soup = BeautifulSoup(res.content, "html.parser")
    title = soup.find_all("h2", class_= "entry-title in-overview")

    # get text and make as list
    length = len(title)
    myList = [title[i].get_text() for i in range(length)]
    
    #convert list as dataframe 
    df = pd.DataFrame(myList)
    df = df.rename(columns={0: get})
    
    return df
```

```python
myList = ["advanced-setups","ahtya","geometry-nodes","kinefx-101","particles","pdg-101","rendering-101","ue-blueprints-101","vellum-101","vex-101","volumes-101","patreon-specials"]
myDfs = [crawlTitle(i) for i in myList]
```

```python
merged_df = pd.concat([df.stack() for df in myDfs], axis=0).unstack()
merged_df
```

```python
merged_df.to_csv("Entagma.csv")
```
